```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Membaca data dari file CSV
df = pd.read_csv('/content/data_penjualan.csv')

# Membuat palet warna yang lebih profesional (biru tua dan hijau tua)
colors = ['#1f77b4', '#2ca02c']

# Menggunakan style grid yang minimalis
sns.set(style="whitegrid")

# Membuat figure dan axis
plt.figure(figsize=(12,6))

# Plot untuk Produk Terjual dengan garis tren profesional
sns.lineplot(x='Bulan', y='Produk Terjual', data=df, marker='o', label='Produk Terjual', color=colors[0], linewidth=2)

# Plot untuk Pendapatan dengan garis tren profesional
sns.lineplot(x='Bulan', y='Pendapatan (IDR)', data=df, marker='o', label='Pendapatan (IDR)', color=colors[1], linewidth=2)

# Menambahkan judul dengan tone yang jelas dan formal
plt.title('Eksplorasi Tren Penjualan Produk dan Pendapatan Bulanan', fontsize=16, fontweight='bold')

# Menambahkan label sumbu dengan font yang bersih dan profesional
plt.xlabel('Bulan', fontsize=12, fontweight='bold')
plt.ylabel('Jumlah Produk Terjual & Pendapatan (IDR)', fontsize=12, fontweight='bold')

# Menambahkan legenda dengan warna yang sesuai dan format rapi
plt.legend(title="Legenda", title_fontsize='13', fontsize='11')

# Menambahkan grid untuk memudahkan eksplorasi data dengan tampilan minimalis
plt.grid(True, which='major', linestyle='--', linewidth=0.5)

# Menghapus spines (garis batas grafik) agar terlihat lebih bersih
sns.despine()

# Menampilkan plot
plt.show()
```
